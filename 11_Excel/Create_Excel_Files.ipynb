```python
from openpyxl import Workbook
from openpyxl.styles import Font
from openpyxl.styles.borders import Border, Side
from openpyxl.styles import PatternFill, Alignment
import datetime
import random

# Die folgenden imports stammen aus der Datei ../00_common/common.py

import sys
sys.path.append("../00_common")
import common

def get_data():
    # define colors
    r = random.choice(["00", "80", "FF"])
    g = random.choice(["00", "80", "FF"])
    if r != "00" and g != "00":
        b = "00"
    else:
        b = random.choice(["08", "FF"])
    color1 = r + g + b
    if r == "FF":
        r = "00"
    elif r == "00":
        r = "FF"
    if g == "FF":
        g = "00"
    elif g == "00":
        g = "FF"
    if b == "FF":
        b = "00"
    elif b == "00":
        b = "FF"
    color2 = r + g + b
    colors = (color1, color2)            
            
    return colors

def make_worksheet_1(wb, matr):
    global colors
    wb.create_sheet("1) Corona")
    ws = wb["1) Corona"]

    l_bundesl = ["Schleswig-Holstein (1)", "Hamburg (2)", "Niedersachsen (3)", "Bremen (4)", "NRW (5)", "Hessen (6)", \
                 "Rheinland-Pfalz (7)", "Baden-Württemberg (8)", "Bayern (9)", "Saarland (10)", "Berlin (11)", "Brandenburg (12)", \
                 "Mecklenburg-Vorpommern (13)", "Sachsen (14)", "Sachsen-Anhalt (15)", "Thüringen (16)"]

    bundesl = l_bundesl[random.randint(0, 15)]
    year = random.choice([2020, 2021, 2022, 2023])

    ws.cell(row = 1, column = 10).value = bundesl
    ws.cell(row = 2, column = 10).value = year

    #ws.cell(row = 1, column = 1).fill = PatternFill(start_color=colors[0], end_color=colors[0], fill_type="solid")
    #ws.cell(row = 1, column = 1).value = colors[0]
    #ws.cell(row = 1, column = 2).fill = PatternFill(start_color=colors[1], end_color=colors[1], fill_type="solid")
    #ws.cell(row = 1, column = 2).value = colors[1]
    

    # Hash and hide the matrikelnumber
    seed = "Hello"
    val = hash(str(matr) + seed) % 10000000000
    ws["BB3"].value = val
    ws["BB3"].font = Font(color="FFFFFF")
    return wb

def make_worksheet_2(wb):
    global colors
    wb.create_sheet("2) Daten")
    ws = wb["2) Daten"]

    separator = random.choice([",", ";", ":", "/"])
    l = ["Datum", "A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "K", "M", "N", "O", "Bewertung", "Kategorie"]
    ws.append([separator.join(l)])

    year = random.randint(2010, 2023)
    startdate = "01-01-" + str(year)
    enddate = "01-01-" + str(year + 1)
    
    start = datetime.datetime.strptime(startdate, "%d-%m-%Y")
    end = datetime.datetime.strptime(enddate, "%d-%m-%Y")
    date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

    for date in date_generated:
        l = [str(date.strftime("%d-%m-%Y"))]
        for _ in range(15):
            l.append(str(random.randint(0, 300)))
        l.append(str(random.randint(0,100) / 10))
        l.append(random.choice(["X", "Y", "Z"]))
        ws.append([separator.join(l)])

    ws.cell(row = 1, column = 20).value = "Farbskala:"
    ws.cell(row = 1, column = 21).value = random.choice(["Grün-Gelb-Rot", "Rot-Gelb-Grün", "Grün-Weiß-Rot", "Rot-Weiß-Grün", \
                                                         "Blau-Weiß-Rot", "Rot-Weiß-Blau", "Grün-Weiß", "Weiß-Grün", "Rot-Weiß", "Weiß-Rot"])
    ws.cell(row = 2, column = 20).value = "Symbolsätze:"
    ws.cell(row = 2, column = 21).value = random.choice(["3 Pfeile (farbig)", "3 Ampeln (ohne Rand)", "3 Symbole (mit Kreis)", "3 Kennzeichen"])
    ws.cell(row = 3, column = 20).value = "Kategorie:"
    ws.cell(row = 3, column = 21).value = random.choice(["X", "Y", "Z"])

    return wb

def make_worksheet_3(wb):
    global colors
    wb.create_sheet("3) Daten")
    ws = wb["3) Daten"]

def make_workbook(filename, matr):
    wb = Workbook()
    wb = make_worksheet_1(wb, matr)
    wb = make_worksheet_2(wb) 
    #wb = make_worksheet_3(wb) 
    #wb = make_worksheet_4(wb)
    del wb["Sheet"]
    return wb

# Main Program
l_matr = common.get_Matrikelnummern("Matrikelnummern.csv")
colors = get_data()
common.create_work_books_in_Excel_Files(l_matr, make_workbook)
#common.create_work_books_in_Abgaben(l_matr, make_workbook)



```
