```python
from openpyxl import load_workbook
from openpyxl import Workbook
from openpyxl.styles import NamedStyle, Font, PatternFill, Alignment
import os

# Die folgenden imports stammen aus der Datei ../00_common/common.py

import sys
sys.path.append("../00_common")
import common

def check_fingerprint(wb, matr):
    ws = wb.worksheets[0]
    if ws["BB3"].value == str(matr) or ws["BA3"].value == str(matr) or ws["AZ3"].value == str(matr):
        return "OK"
    else:
        return "NOK"

def check_arrayformula(wb, matr):
    try:
        response = "        "
        ws = wb.worksheets[0]
        if ws.array_formulae != {}:
            response = response[:1] + "1" + response[2:]
        
        ws = wb.worksheets[1]
        if ws.array_formulae != {}:
            response = response[:3] + "2" + response[4:]

        ws = wb.worksheets[2]
        if ws.array_formulae != {}:
            response = response[:5] + "3" + response[6:]

        return response
    except:
        print("FAIL:", matr, "check Matrix_Formula")
        return "NOK"

def check_exercise_1_1(wb, matr):
    try:
        ws = wb.worksheets[0]
        points = 0
        # Check formatting of cells D1:F1
        if ws["D1"].fill.start_color.index == "FF00B1AC" and ws["E1"].fill.start_color.index == "FF00B1AC" and ws["F1"].fill.start_color.index == "FF00B1AC":
            points += 1
        return points
    except:
        print("FAIL:", matr, "check_exercise_1")
        return "NOK"

def check_exercise_1_2(wb_val, matr):
    try:
        ws = wb_val.worksheets[0]
        points = 0
        # Check content of E-Mail in H10, H20, H30
        for row in [10, 20, 30]:
            fname = ws["A" + str(row)].value
            name = ws["B" + str(row)].value
            mail = ws["H" + str(row)].value
            if mail == fname.lower() + "." + name.lower() + "@alumni.fh-aachen.de":
                points += 1
        return points
    except:
        print("FAIL:", matr, "check_exercise_1_2")
        return "NOK"

def check_exercise_1_3(wb, matr):
    try:
        ws = wb.worksheets[0]
        points = 0
        # Check if formula with Text-Formula is in cell H40
        if ws["H40"].value != None:
            formula = ws["H40"].value
            if formula[0] == "=" and "A40" in formula and "B40" in formula:
                points += 4
        return points
    except:
        print("FAIL:", matr, "check_exercise_1_3")
        return "NOK"

def check_exercise_1_4(wb_val, matr):
    try:
        ws = wb_val.worksheets[0]
        points = 0
        # Check if value in J2 is double-mail in Column H
        if ws["J2"].value != None:
            double_mail = ws["J2"].value
            l_mail = []
            row_num = 2
            while ws.cell(row = row_num, column = 8).value != None:
                l_mail.append(ws.cell(row = row_num, column = 8).value)
                row_num += 1
            if l_mail.count(double_mail) == 2:
                points += 4
        return points
    except:
        print("FAIL:", matr, "check_exercise_1_4")
        return "NOK"

def check_exercise_2_1(wb_val, matr):
    try:
        ws = wb_val.worksheets[1]
        points = 0
        # Check if B50 is Firstname from E-Mail BS and C50 ist name from email
        if ws["B50"].value != None:
            fname = ws["B50"].value
            email = ws["A50"].value
            ind_dot = email.find(".")
            email_fname = email[:ind_dot]
            if fname.lower() == email_fname:
                points += 4
        if ws["C50"].value != None and len(str(ws["C50"].value)) > 1:
            name = ws["C50"].value
            email = ws["A50"].value
            ind_dot = email.find(".")
            ind_at = email.find("@")
            email_name = email[ind_dot + 1:ind_at]
            if name.lower() == email_name:
                points += 4
            
        return points
    except:
        print("FAIL:", matr, "check_exercise_2_1")
        return "NOK"

def check_exercise_2_2(wb, matr):
    try:
        ws = wb.worksheets[1]
        points = 0
        # Check if formula with Text-Formula is in cell B35 and C35
        if ws["B35"].value != None:
            formula = ws["B35"].value
            if formula[0] == "=" and "35" in formula:
                points += 2
        if ws["C35"].value != None:
            formula = ws["C35"].value
            if formula[0] == "=" and "35" in formula:
                points += 2
        return points
    except:
        print("FAIL:", matr, "check_exercise_2_2")
        return "NOK"

def check_exercise_3_1(wb_val, matr):
    try:
        ws = wb_val.worksheets[2]
        points = 0
        # Check if D60 is contains correct tel
        if ws["D60"].value != None:
            new_tel = ws["D60"].value
            tel = ws["C60"].value
            ind_slash = tel.find("/")
            tel1 = tel[1:ind_slash - 1]
            tel2 = tel[ind_slash + 2:]
            tel = "+49 (" + tel1 + ") " + tel2
            ind_bra = tel.find(")")
            ind_bra_new = new_tel.find(")")
            if tel[:ind_bra] == new_tel[:ind_bra_new]:
                points += 3
            if tel[ind_bra:] == new_tel[ind_bra_new:]:
                points += 3
            
        return points
    except:
        print("FAIL:", matr, "check_exercise_3_1")
        return "NOK"

def check_exercise_3_2(wb, matr):
    try:
        ws = wb.worksheets[2]
        points = 0
        # Check if formula with Text-Formula is in cell D50
        if ws["D50"].value != None:
            formula = ws["D50"].value
            if formula[0] == "=" and "50" in formula:
                points += 3
        return points
    except:
        print("FAIL:", matr, "check_exercise_3_2")
        return "NOK"

def check_workbook(matr, wb, wb_val):
    points = [matr]
    points.append(check_fingerprint(wb, matr))
    points.append(check_arrayformula(wb, matr))
    points.append(check_exercise_1_1(wb, matr))
    points.append(check_exercise_1_2(wb_val, matr))
    points.append(check_exercise_1_3(wb, matr))
    points.append(check_exercise_1_4(wb_val, matr))
    points.append(check_exercise_2_1(wb_val, matr))
    points.append(check_exercise_2_2(wb, matr))
    points.append(check_exercise_3_1(wb_val, matr))
    points.append(check_exercise_3_2(wb, matr))
    
    return points



l_points = common.check_ExcelFiles_in_Abgaben(check_workbook)

for points in l_points:
    print(points)
    
    header = ["MatrNr", "Fingerprint", "Matrix-F", "1 Format", "1 Values", "1 Formula", "1 Double Mail", "2 Value", "2 Formula", "3 Value", "3 Formula"] 
common.result_into_file(l_points, header)
```
