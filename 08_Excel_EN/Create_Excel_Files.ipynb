```python
from openpyxl import Workbook
from openpyxl.styles import Font
from openpyxl.styles.borders import Border, Side
from openpyxl.styles import PatternFill, Alignment
import random

# Die folgenden imports stammen aus der Datei ../00_common/common.py

import sys
sys.path.append("../00_common")
import common

def get_data():
    l_customers = []
    with open("persons_1000.csv", "r", encoding='utf-8-sig') as file:
        for line in file:
            line = line.strip()
            line = line.split(";")
            line = tuple(line)
            l_customers.append(line)
    
    l_ikea_names = []
    with open("ikea_like_names.csv", "r", encoding="utf-8-sig") as file:
        for line in file:
            line = line.strip()
            l_ikea_names.append(line)

    l_emails = []
    with open("names_2000.csv", "r", encoding="utf-8-sig") as file:
        for line in file:
            line = line.strip()
            fname, name = line.split(";")
            email = fname + "." + name + "@" + random.choice(["yahoo", "gmail", "outlook", "hotmail", "aol", "web", "gmx"]) + \
                         "." + random.choice(["de", "nl", "be", "uk", "it", "es", "pt", "ch", "se", "gr", "no", "ir", "pl"])
            l_emails.append(email)
            
    return l_customers, l_ikea_names, l_emails

def make_worksheet_1(wb, matr):
    global l_customers
    global l_ikea_names
    wb.create_sheet("1) Invoice")
    ws = wb["1) Invoice"]
    
    # Draw a border around the invoice
    for row_num in range(2, 42):
        ws.cell(row=row_num, column=2).border = Border(left=Side(style='medium'))
        ws.cell(row=row_num, column=8).border = Border(right=Side(style='medium'))
    for col_num in range(2, 9):
        ws.cell(row=2, column=col_num).border = Border(top=Side(style='medium'))
        ws.cell(row=41, column=col_num).border = Border(bottom=Side(style='medium'))
        ws.cell(row=18, column=col_num).border = Border(bottom=Side(style='thin'))
        ws.cell(row=38, column=col_num).border = Border(bottom=Side(style='thin'))
        ws.cell(row=29, column=col_num).border = Border(bottom=Side(style='thin'))
    ws.cell(row=2, column=2).border = Border(left=Side(style="medium"), top=Side(style="medium"))
    ws.cell(row=2, column=8).border = Border(right=Side(style="medium"), top=Side(style="medium"))
    ws.cell(row=41, column=2).border = Border(left=Side(style="medium"), bottom=Side(style="medium"))
    ws.cell(row=41, column=8).border = Border(right=Side(style="medium"), bottom=Side(style="medium"))
    ws.cell(row=18, column=2).border = Border(left=Side(style="medium"), bottom=Side(style="thin"))
    ws.cell(row=18, column=8).border = Border(right=Side(style="medium"), bottom=Side(style="thin"))
    ws.cell(row=29, column=2).border = Border(left=Side(style="medium"), bottom=Side(style="thin"))
    ws.cell(row=29, column=8).border = Border(right=Side(style="medium"), bottom=Side(style="thin"))
    ws.cell(row=38, column=2).border = Border(left=Side(style="medium"), bottom=Side(style="thin"))
    ws.cell(row=38, column=8).border = Border(right=Side(style="medium"), bottom=Side(style="thin"))
    ws.column_dimensions["G"].width = 15
    ws.column_dimensions["F"].width = 15
    
    # Fill parts of the invoice with constants
    ws["G9"].value = "OGIA Gmbh"
    ws["G10"].value = "ABC-Street 123"
    ws["G11"].value = "98765 Somewhere"
    ws["G14"].value = "Customer-ID:"
    ws["G15"].value = "Invoice-No:"
    ws["G16"].value = "Date"
    ws["B19"].value = "Number"
    ws["C19"].value = "Prod-No."
    ws["D19"].value = "Product"
    ws["E19"].value = "Number"
    ws["F19"].value = "Price"
    ws["G19"].value = "Total"
    ws["E30"].value = "Intermediate"
    ws["E31"].value = "Discount"
    ws["E32"].value = "Sum"
    ws["E33"].value = "VAT"
    ws["E34"].value = "To be paid:"
    ws["B39"].value = "Payment Due Date:"
    ws["B40"].value = "IBAN:   DE12 3456 7890 1234 5678 90"
    ws["B41"].value = "BIC:    WEKSDSC"

    # Little polishing
    for col_num in range(2, 8):
        ws.cell(row = 19, column = col_num).font = Font(bold=True)

    # Mark cells to be filled with formulas with background-color #D9D9D9
    for row_num in range(9, 12):
        ws.cell(row = row_num, column = 3).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
    ws.cell(row = 16, column = 8).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
    for row_num in range(20, 30):
        ws.cell(row = row_num, column = 2).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
        ws.cell(row = row_num, column = 4).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
        ws.cell(row = row_num, column = 6).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
        ws.cell(row = row_num, column = 7).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
    for row_num in range(30, 35):
        ws.cell(row = row_num, column = 7).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
    ws.cell(row = 31, column = 6).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
    ws.cell(row = 39, column = 4).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")

    # Mark cells to be filled with constants with background-color #FCD5B4
    ws.cell(row = 14, column = 8).fill = PatternFill(start_color="FCD5B4", fill_type = "solid")
    ws.cell(row = 15, column = 8).fill = PatternFill(start_color="FCD5B4", fill_type = "solid")
    for row_num in range(20, 30):
        ws.cell(row = row_num, column = 3).fill = PatternFill(start_color="FCD5B4", fill_type = "solid")
        ws.cell(row = row_num, column = 5).fill = PatternFill(start_color="FCD5B4", fill_type = "solid")
    
    # Create list of Customers
    num_customers = random.randint(550, 650)
    sublist_customers = random.sample(l_customers, num_customers)

    ws["J1"].value = "Customer-DB"
    ws.cell(row = 2, column = 10).value = "Customer-ID"
    ws.cell(row = 2, column = 11).value = "Firstname"
    ws.cell(row = 2, column = 12).value = "Name"
    ws.cell(row = 2, column = 13).value = "Street"
    ws.cell(row = 2, column = 14).value = "City"
    ws.cell(row = 2, column = 15).value = "Phone"
    ws.cell(row = 2, column = 16).value = "Discount"
    
    row_num = 3
    cust_num = 1
    for customer in sublist_customers:
        fname, name, street, no, zip, city, tel = customer
        street = street + " " + no
        city = zip + " " + city
        discount = random.choice([0, 0.12, 0.15, 0, 0])
        ws.cell(row = row_num, column = 10).value = cust_num
        ws.cell(row = row_num, column = 11).value = fname
        ws.cell(row = row_num, column = 12).value = name
        ws.cell(row = row_num, column = 13).value = street
        ws.cell(row = row_num, column = 14).value = city
        ws.cell(row = row_num, column = 15).value = tel
        ws.cell(row = row_num, column = 16).value = discount
        row_num += 1
        cust_num += 1
    
    # Create a list of Products
    num_products = random.randint(60, 80)
    sublist_products = random.sample(l_ikea_names, num_products)

    ws["R1"].value = "Produkt-DB"
    ws.cell(row = 2, column = 18).value = "Product-No"
    ws.cell(row = 2, column = 19).value = "Product"
    ws.cell(row = 2, column = 20).value = "Price"

    row_num = 3
    prod_num = 1
    for product in sublist_products:
        ws.cell(row = row_num, column = 18).value = prod_num
        ws.cell(row = row_num, column = 19).value = product
        price = random.choice([19, 29, 39, 49, 79, 89, 99]) + random.choice([0, 0, 0, 100, 200])
        ws.cell(row = row_num, column = 20).value = price
        row_num += 1
        prod_num += 1   

    # Hash and hide the matrikelnumber
    seed = "abc"
    val = hash(str(matr) + seed) % 10000000000
    ws["BB3"].value = val
    ws["BB3"].font = Font(color="FFFFFF")
    return wb

def make_worksheet_2(wb):
    wb.create_sheet("2) Project Plan")
    ws = wb["2) Project Plan"]

    ws["A1"].value = "Project Plan XYZ"
    
    # Add weeks
    num_wks = random.randint(12, 18)
    start_wk = random.randint(20, 30)
    col = 2
    for wk in range(start_wk, start_wk + num_wks):
        ws.cell(row = 2, column = col).value = wk
        col += 1

    # Add Workpackages
    num_wps = random.randint(20, 30)
    for wp in range(1, num_wps):
        ws.cell(row = wp + 2, column = 1).value = wp

    for col in range(2, num_wks +2):
        ws.cell(row = 3, column = col).value = "PM"
        ws.cell(row = 3, column = col).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
        ws.cell(row = 3, column = col).alignment = Alignment(horizontal='center')

    l_wp = []
    for i in range(num_wps - 2):
        wp_begin = random.randint(1, num_wks - 2)
        l_wp.append(wp_begin)
    l_wp.sort()
    row_num = 4
    for wp in l_wp:
        wp_length = random.randint(2, 8)
        wp_end = min(wp + wp_length, num_wks)
        empl = random.choice(["SW", "PR", "PR", "PR", "TE", "TE"])
        for col in range(wp + 2, wp_end + 2):
            ws.cell(row = row_num, column = col).value = empl
            ws.cell(row = row_num, column = col).fill = PatternFill(start_color="D9D9D9", fill_type = "solid")
            ws.cell(row = row_num, column = col).alignment = Alignment(horizontal='center')
        row_num += 1

    # Add Cost rows
    ws.cell(row = num_wps + 3, column = 1).value = "PM"
    ws.cell(row = num_wps + 4, column = 1).value = "SW"
    ws.cell(row = num_wps + 5, column = 1).value = "PR"
    ws.cell(row = num_wps + 6, column = 1).value = "TE"
    ws.cell(row = num_wps + 7, column = 1).value = "Cost/Wk"
    ws.cell(row = num_wps + 8, column = 1).value = "Cost acc."

    # Add table with employer costs
    ws.cell(row = 1, column = num_wks + 3).value = "Cost per Employee and Week"
    ws.cell(row = 2, column = num_wks + 3).value = "Project Manager"
    ws.cell(row = 2, column = num_wks + 4).value = "PM"
    ws.cell(row = 2, column = num_wks + 5).value = random.randint(35, 50) * 100
    ws.cell(row = 3, column = num_wks + 3).value = "SW-Architect"
    ws.cell(row = 3, column = num_wks + 4).value = "SW"
    ws.cell(row = 3, column = num_wks + 5).value = random.randint(35, 50) * 100
    ws.cell(row = 4, column = num_wks + 3).value = "Programmer"
    ws.cell(row = 4, column = num_wks + 4).value = "PR"
    ws.cell(row = 4, column = num_wks + 5).value = random.randint(25, 40) * 100
    ws.cell(row = 5, column = num_wks + 3).value = "Tester"
    ws.cell(row = 5, column = num_wks + 4).value = "TE"
    ws.cell(row = 5, column = num_wks + 5).value = random.randint(30, 40) * 100

    return wb

def make_worksheet_3(wb):
    global l_emails
    wb.create_sheet("3) E-Mails")
    ws = wb["3) E-Mails"]

    num_emails = random.randint(1300, 1500)
    subl_emails = random.sample(l_emails, num_emails)

    tlds = ["de", "nl", "be", "uk", "it", "es", "pt", "ch", "se", "gr", "no", "ir", "pl"]
    tlds.sort()
    line = ["E-Mail", "Domain", "Top-Level-Domain", "", ""] + tlds + ["Sum"]
    ws.append(line)
    ws.column_dimensions["A"].width = 25
    ws.column_dimensions["B"].width = 15
    ws.column_dimensions["C"].width = 15

    for email in subl_emails:
        ws.append([email])
    
    l_domains = ["yahoo", "gmail", "outlook", "hotmail", "aol", "web", "gmx"]
    l_domains.sort()
    l_domains += ["Sum"]

    row_num = 2
    for domain in l_domains:
        ws.cell(row = row_num, column = 5).value = domain
        row_num += 1

    return wb

def make_worksheet_4(wb):
    wb.create_sheet("4) Logic Brain Teaser")
    ws = wb["4) Logic Brain Teaser"]

    ws.append(["Logic Brain Teaser"])

    ws.append([False, False, True, True, "", False, False, True, True])
    ws.append([False, True, False, True, "", False, True, False, True])
    ws.append([False, False, False, False])   
    ws.append([False, False, False, True])
    ws.append([False, False, True, False])
    ws.append([False, False, True, True])
    ws.append([False, True, False, False])
    ws.append([False, True, False, True])
    ws.append([False, True, True, False])
    ws.append([False, True, True, True])
    ws.append([True, False, False, False])   
    ws.append([True, False, False, True])
    ws.append([True, False, True, False])
    ws.append([True, False, True, True])
    ws.append([True, True, False, False])
    ws.append([True, True, False, True])
    ws.append([True, True, True, False])
    ws.append([True, True, True, True])

    ws.cell(row = 5, column = 6).value = "=AND(F2,F3)"
    ws.cell(row = 5, column = 7).value = "=AND(G2,G3)"
    ws.cell(row = 5, column = 8).value = "=AND(H2,H3)"
    ws.cell(row = 5, column = 9).value = "=AND(I2,I3)"
    ws.cell(row = 11, column =6).value = "=OR(F2,F3)"
    ws.cell(row = 11, column = 7).value = "=OR(G2,G3)"
    ws.cell(row = 11, column = 8).value = "=OR(H2,H3)"
    ws.cell(row = 11, column = 9).value = "=OR(I2,I3)"
    ws.cell(row = 9, column = 6).value = "=F3"
    ws.cell(row = 9, column = 7).value = "=G3"
    ws.cell(row = 9, column = 8).value = "=H3"
    ws.cell(row = 9, column = 9).value = "=I3"
    ws.cell(row = 19, column = 6).value = "=OR(F2,NOT(F2))"
    ws.cell(row = 19, column = 7).value = "=OR(G2,NOT(G2))"
    ws.cell(row = 19, column = 8).value = "=OR(H2,NOT(H2))"
    ws.cell(row = 19, column = 9).value = "=OR(I2,NOT(I2))"

    # Some formatting
    ws["A1"].font = Font(size = 16, bold = True)
    for row_num in range(2, 4):
        for col_num in range(1, 10):
            if col_num != 5:
                ws.cell(row = row_num, column = col_num).fill = PatternFill(start_color="FF00B1AC", fill_type = "solid")
                ws.cell(row = row_num, column = col_num).font = Font(color = "FFFFFF")

    return wb
    
def make_workbook(filename, matr):
    wb = Workbook()
    wb = make_worksheet_1(wb, matr)
    wb = make_worksheet_2(wb) 
    wb = make_worksheet_3(wb) 
    wb = make_worksheet_4(wb)
    del wb["Sheet"]
    return wb

# Main Program
l_matr = common.get_Matrikelnummern("Matrikelnummern.csv")
l_customers, l_ikea_names, l_emails = get_data()
common.create_work_books_in_Excel_Files(l_matr, make_workbook)
#common.create_work_books_in_Abgaben(l_matr, make_workbook)
```
