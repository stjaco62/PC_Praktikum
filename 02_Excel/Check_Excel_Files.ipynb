```python
from openpyxl import load_workbook
from openpyxl import Workbook
from openpyxl.styles import NamedStyle, Font, PatternFill, Alignment
import os

# Die folgenden imports stammen aus der Datei ../00_common/common.py

import sys
sys.path.append("../00_common")
import common

def check_fingerprint(wb, matr):
    try:
        ws = wb["1) Formel"]
        if ws["BB3"].value == int(matr):
            return "OK"
        else:
            return "NOK"
    except:
        print("FAIL:", matr, "check_fingerprint")
        return "NOK"

def check_exercise_1_1(wb, matr):
    # Check formatting of cells D6 and B10 as examples
    try: 
        ws = wb["1) Formel"]
        points = 0
        # Check if background color is FH-Mint
        if ws["E3"].fill.start_color.index == ws["B10"].fill.start_color.index == "FF00B1AC":
            points += 1
    
        # Check, if the numberformat in cells D7 has one number after the decimal point
        if ws["D7"].number_format == "0.0":
            points += 1
        
        return points
    except:
        print("FAIL:", matr, "exercise 1_1")
        return "NOK"

def check_exercise_1_2(wb, matr):
    # Check if addressing of cells is correct (look for $s)
    try: 
        ws = wb["1) Formel"]
        points = 0
        if ws["C4"].value != None:
            if "C$3" in ws["C4"].value and "$B4" in ws["C4"].value:
                points += 4
        return points
    except:
        print("FAIL:", matr, "exercise 1_2")
        return "NOK"

def check_exercise_1_3(wb_val, matr):
    # Check if the calculations are done correctly (by example D10, J5)
    try:
        ws = wb_val["1) Formel"]
        points = 0
        if ws["B1"] != None:
            formula = ws["B1"].value
            if ws["D10"] != None and ws["D3"] != None and ws["B10"] != None:
                formula = formula.replace("a", str(ws["D3"].value))
                formula = formula.replace("b", str(ws["B10"].value))
                if ws["D10"].value == eval(formula):
                    points += 2
            formula = ws["B1"].value
            if ws["J5"] != None and ws["J3"] != None and ws["B5"] != None:
                formula = formula.replace("a", str(ws["J3"].value))
                formula = formula.replace("b", str(ws["B5"].value))
                if ws["J5"].value == eval(formula):
                    points += 2
        return points
    except:
        print("FAIL:", matr, "exercise 1_3")
        return "NOK"

def check_exercise_2_1(wb, matr):
    # Check formatting of cells B14 and G6 as examples
    try: 
        ws = wb["2) Prüfung"]
        points = 0
        # Check if background color is FH-Mint
        if ws["B14"].fill.start_color.index == ws["G6"].fill.start_color.index == "FF00B1AC":
            points += 1
    
        # Check, if the numberformat in cells D13 has two digits after the decimal point and E1 and I6 have one digit
        if ws["D13"].number_format == "0.00" and ws["E1"].number_format == "0.0" and ws["I6"].number_format == "0.0":
            points += 1
    
        return points
    except:
        print("FAIL:", matr, "exercise 2_1")
        return "NOK"

def check_exercise_2_2(wb, matr):
    # Check if addressing of cells is correct (look for $s)
    try: 
        ws = wb["2) Prüfung"]
        points = 0

        if ws.array_formulae != {}:
            print(matr, ws.array_formulae)
        if ws["C7"].value != None:
            if "$B$2" in ws["C7"].value and "$B$3" in ws["C7"].value and "$B$4" in ws["C7"].value \
            and "$E$1" in ws["C7"].value and "C$6" in ws["C7"].value and "$B7" in ws["C7"].value:
                points = 4
            if "$B$2" in ws["C7"].value and "$B$3" in ws["C7"].value and "$B$4" in ws["C7"].value \
            and "Note_Praktikum" in ws["C7"].value and "C$6" in ws["C7"].value and "$B7" in ws["C7"].value:
                points = 4
        return points
    except:
        print("FAIL:", matr, "exercise 2_2")
        return "NOK"

def check_exercise_2_3(wb_val, matr):
    # Check if the calculations are done correctly (by example D10, J13)
    try: 
        ws = wb_val["2) Prüfung"]
        points = 0
        if ws["B2"].value != None and ws["E1"].value != None:
            s1 = ws["B2"].value * ws["E1"].value
        if ws["B3"].value != None and ws["D6"].value != None:
            s2 = ws["B3"].value * ws["D6"].value
        if ws["B4"].value != None and ws["B10"].value != None:
            s3 = ws["B4"].value * ws["B10"].value
        s = s1 + s2 + s3
        if ws["D10"].value != None:
            if round(s, 2) == round(ws["D10"].value, 2):
                points += 2
        if ws["B2"].value != None and ws["E1"].value != None:
            s1 = ws["B2"].value * ws["E1"].value
        if ws["B3"].value != None and ws["J6"].value != None:
            s2 = ws["B3"].value * ws["J6"].value
        if ws["B4"].value != None and ws["B13"].value != None:
            s3 = ws["B4"].value * ws["B13"].value
        s = s1 + s2 + s3
        if ws["J13"].value != None:
            if round(s, 2) == round(ws["J13"].value, 2):
                points += 2
        return points
    except:
        print("FAIL:", matr, "exercise 2_3")
        return "NOK"

def check_exercise_2_4(wb, matr):
    # Check if the calculations are done correctly (by example D10, J13)
    try:
        ws = wb["2) Prüfung"]
        points = 0
        if "Note_Praktikum" in wb.defined_names:
            points += 1
            if wb.defined_names["Note_Praktikum"].attr_text == "'2) Prüfung'!$E$1":
                points += 1
        return points
    except:
        print("FAIL:", matr, "exercise 2_4")
        return "NOK"

def check_workbook(matr, wb, wb_val):
    points = [matr]
    points.append(check_fingerprint(wb, matr))
    points.append(check_exercise_1_1(wb, matr))
    points.append(check_exercise_1_2(wb, matr))
    points.append(check_exercise_1_3(wb_val, matr))
    points.append(check_exercise_2_1(wb, matr))
    points.append(check_exercise_2_2(wb, matr))
    points.append(check_exercise_2_3(wb_val, matr))
    points.append(check_exercise_2_4(wb, matr))
    return points



l_points = common.check_ExcelFiles_in_Abgaben(check_workbook)
header = ["MatrNr", "Fingerprint", "1 Format", "1 Bezüge", "1 Formel", "2 Format", "2 Bezüge", "2 Formel", "2 Name"] 
common.result_into_file(l_points, header)
```
